```python
class Creature:
    """Base class for all creatures, 
    defining common 
    attributes and behaviors."""
    
    def __init__(self, name, species, power_level, speed, habitat, special_ability, weakness):
        """Initializes an instatiation method 
        with its parameters."""
        self.name = name
        # Assigning property self.name to parameter name
        self.species = species
        # Assigning property to parameter species
        self.power_level = power_level 
        # Assigning property self.power_level to parameter power_level
        self.speed = speed 
        # Assigning property self.speed to parameter speed
        self.habitat = habitat 
        # Assigning property self.habitat to parameter habitat
        self.special_ability = special_ability 
        # Assigning property self.special_ability to parameter special_ability
        self.weakness = weakness  
        # Assigning property self.weakness to parameter weakness

    def describe(self):
        """Prints out the details of the creature."""
        print(f"Creature: {self.name}")
        print(f"Species: {self.species}")
        print(f"Power Level: {self.power_level}")
        print(f"Speed: {self.speed}")
        print(f"Habitat: {self.habitat}")
        print(f"Special Ability: {self.special_ability}")
        print(f"Weakness: {self.weakness}")
        print("-" * 30)

    def use_ability(self):
        """Defining a method 
        called use_ability.
        The method prints properties name and special_ability"""
        print(f"{self.name} uses {self.special_ability}!")


#A subclass of Creature representing a Dragon, capable of flying Inheriting from class Creature.
class Dragon(Creature):
    
    def fly(self):
        """Enables the dragon to fly."""
        print(f"{self.name} spreads its wings and takes off into the sky!")

# A subclass of Dragon representing an Elf, with stealth abilities and inheriting from class Dragon
class Elf(Dragon):
    
    def stealth(self):
        """Enables the elf to blend into the shadows."""
        print(f"{self.name} blends into the shadows, becoming nearly invisible.")

# A subclass of Elf representing a Mermaid, capable of swimming effortlessly inherinting from Elf.
class Mermaid(Elf):
   
    
    def swim(self):
        """ A method called swim printing
        property name"""
        print(f"{self.name} glides effortlessly through the waves.")


class Golem(Mermaid):
    """A subclass of Mermaid representing a Golem, with immense strength and resilience."""
    
    def __init__(self, name, power_level, speed):
        """Initializes a Golem with predefined species, habitat, abilities, and weaknesses."""
        super().__init__(name, "Golem", power_level, speed, "Forests", "Archery Mastery", "Dark Magic")
    
    def smash(self):
        """Enables the golem to perform a powerful smash attack."""
        print(f"{self.name} slams the ground, causing a small earthquake!")


# Creating instances of the creatures
drake = Dragon("Drake", 90, 45, "40km/hr", "forest", "sky diving", "prune")
legolas = Elf("Legolas", 80, 100, "60km/hr", "desert", "fast runner", "can't fly")
ariel = Mermaid("Ariel", 90, 45, "90km/hr", "water", "swimming", "net")
rocky = Golem("Rocky", 90, 45)

# Performing actions
drake.describe()
drake.use_ability()
drake.fly()

legolas.describe()
legolas.use_ability()
legolas.stealth()

ariel.describe()
ariel.use_ability()
ariel.swim()

rocky.describe()
rocky.use_ability()
rocky.smash()

""I"Assigns values to all attributes of the Golem object including those inherited from the Mermaid class """
ayikoru alima
1:22â€¯PM
"""Prints out the Golem object's name and how it achieves the smashing action """
```
